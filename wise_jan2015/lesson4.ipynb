#Challenge 2

Rewrite the Code that sums positive and negative numbers in a list using in place operators

```python
def posNeg(numList):
    '''this function calculates the totals for all negative numbers and all positive numbers in a list
    separately. 
    Inputs: a list of numbers
    outputs: a list containing [sumPositiveNumbers, SumNegativeNumbers]'''
    totalPos = 0
    totalNeg = 0
    for nums in numList:
        if nums >= 0:
            totalPos = totalPos + nums
        else:
            totalNeg = totalNeg + nums
    return [totalPos, totalNeg ]
    
numbers=[1,2,3,-4,-3,2,-9]
posNeg(numbers)
```

```python
def make_heatmap(values, low_color, mid_color, high_color, low_band, high_band, block_size):
    ''' this function will make a 3 color heatmap that visually illustrates the temporal spread of the data 
    in a given dataset'''
    width, height=values.shape
    result = ImageGrid(width,height, block_size=block_size)
    center = values.mean()
    for x in range(width):
        for y in range(height):
            if values[x,y] < (center*low_band):
                result[x,y] = low_color
            elif values[x,y] > (center*high_band):
                result[x,y] = high_color
            else:
                result[x,y]= mid_color
    return result           
            
```

### heatmap function with assigned default values

```python
def make_heatmap(values,
                 low_band=0.5, high_band=1.5,
                 low_color=colors['Gray'], mid_color=colors['YellowGreen'], high_color=colors['SpringGreen'],
                 block_size=5):
    '''Make a 3-colored heatmap from a 2D array of data.
    Default color scheme is gray to green.'''
    width, height = values.shape
    result = ImageGrid(width, height, block_size=block_size)
    center = values.mean()
    for x in range(width):
        for y in range(height):
            if values[x, y] < low_band * center:
                result[x, y] = low_color
            elif values[x, y] > high_band * center:
                result[x, y] = high_color
            else:
                result[x, y] = mid_color
    return result
```
